```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math

y0=[0,0.1]
t=np.linspace(0, 5, 100)

g=9.81;
Fb=2;
Fs=0.5
m=0.05
Fk=1
c1=0.2

def vib(y, t, g, Fb, Fs, m, Fk, c1):
    A1, A2=y
    dydt=[-g+(Fb-Fs)/m-c1/m*A1+c1/m*A2-Fk/m, -g+c1/m*A1-c1/m*A2+Fk/m]
    return dydt


sol = odeint(vib, y0, t, args=(g, Fb, Fs, m, Fk, c1))

plt.plot(t, sol[:, 0], 'b', label='A1')
plt.plot(t, sol[:, 1], 'g', label='A2')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
